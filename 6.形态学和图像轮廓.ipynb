# 开运算

```python
import cv2
import numpy as np

# 开运算 = 腐蚀 -> 膨胀
# 开运算提供了另一种去除噪声的思路

img = cv2.imread('./fushi.png')
img2 = cv2.imread('./hujiao.png')

kernel =  cv2.getStructuringElement(cv2.MORPH_RECT,(5,5))
# # 先腐蚀
# dst = cv2.erode(img2,kernel,iterations=1)
# # 膨胀
# dst1 = cv2.dilate(dst,kernel,iterations=1)

# 调用 opencv 提供的api
dst = cv2.morphologyEx(img2,cv2.MORPH_OPEN,kernel,iterations=1)

cv2.imshow('img',np.hstack((img2,dst)))
cv2.waitKey(0)
cv2.destoryAllWindows()
```

# 闭运算

```python
# 闭运算用于消除图形内部的噪声
import cv2
import numpy as np

img = cv2.imread('./j.png')

kernel =  cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))

# 调用 opencv 提供的api
dst = cv2.morphologyEx(img,cv2.MORPH_CLOSE,kernel,iterations=6)

cv2.imshow('img',np.hstack((img,dst)))
cv2.waitKey(0)
cv2.destoryAllWindows()
```

# 形态学梯度

```python
# 形态学梯度 = 原图 - 腐蚀
import cv2
import numpy as np

img = cv2.imread('j.png')

kernel =  cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))

dst = cv2.morphologyEx(img,cv2.MORPH_CLOSE,kernel,iterations=6)
# 调用 opencv 提供的api
dst = cv2.morphologyEx(dst,cv2.MORPH_GRADIENT,kernel,iterations=1)

cv2.imshow('img',np.hstack((img,dst)))
cv2.waitKey(0)
cv2.destoryAllWindows()
```

# 绘制轮廓

```python
import cv2
import numpy as np

img = cv2.imread('./j.png')

gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

# 二值化
thresh ,binary = cv2.threshold(gray,150,255,cv2.THRESH_BINARY)

# 查找轮廓，返回轮廓和层级
contours,hierarchy = cv2.findContours(binary,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

# 绘制轮廓会直接修改原图
img_copy = img
cv2.drawContours(img_copy,contours, -1, (0,0,255), 2)

cv2.imshow('img_copy',img_copy)

cv2.waitKey(0)
cv2.destoryAllWindows()
```

# 多边形逼近

```python
import cv2

img = cv2.imread('./hand.png')

gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

# 二值化
thresh,binary = cv2.threshold(gray,150,255,cv2.THRESH_BINARY)

contours,hierarchy = cv2.findContours(binary,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

# 绘制轮廓会直接修改原图
img_copy = img.copy()
cv2.drawContours(img_copy,contours,0,(0,0,255),2)

# 使用多边形逼近，近似模拟手的轮廓
approx = cv2.approxPolyDP(contours[0],20,closed=True)
# approx 本质上就是一个轮廓数据
# 所以也用到 绘制轮廓的方法
cv2.drawContours(img_copy,[approx],0,(0,255,0),2)

cv2.imshow('img',img)
cv2.imshow('img_copy',img_copy)
cv2.waitKey(0)
cv2.destoryAllWindows()
```
